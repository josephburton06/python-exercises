```python
import matplotlib.pyplot as plt
```

```python
# Use matplotlib to plot the following equation:
# y=x^2−x+2
# Add an anotation for the point 0, 0, the origin.
```

```python
x = list(range(-20,20))
y = [n ** 2 - n + 2 for n in x]
plt.scatter(x, y)
plt.annotate('Origin', xy = (0,10), xytext = (-2,50), arrowprops={'facecolor': 'red'})
plt.title('y = x^2 - x +2')
plt.show()
```

```python
# Create and label 4 separate plots for the following equations (choose a range for x that makes sense):

# y=√x
# y=x^3
# y=tan(x)
# y=2^x

# You can use functions from the math module to help implement some of the equations above.
```

```python
import math as math
```

```python
# y=√x
x = list(range(0,20))
y = [math.sqrt(n) for n in x]
plt.plot(x,y)
plt.title('y = $sqrt$(x)')
plt.show()
```

```python
# y=x^3
x = list(range(-20,20))
y = [n ** 3 for n in x]
plt.plot(x,y)
plt.title('y = x^3')
plt.show()
```

```python
# y=tan(x)
x = list(range(-20,20))
y = [math.tan(n) for n in x]
plt.plot(x,y)
plt.title('y = tan(x)')
plt.show()
```

```python
# y=2^x
x = list(range(0,10))
y = [(2 ** n) for n in x]
plt.plot(x,y)
plt.title('y = 2^x')
plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
```

```python
# y=√x
x1 = list(range(0,20))
y1 = [math.sqrt(n) for n in x1]
plt.subplot(221)
plt.plot(x1,y1)
plt.suptitle('y = $sqrt$(x)')
plt.show()

# y=x^3
x2 = list(range(-20,20))
y2 = [n ** 3 for n in x2]
plt.subplot(222)
plt.plot(x2,y2)
plt.suptitle('y = x^3')
plt.show()

# y=tan(x)
x3 = list(range(-20,20))
y3 = [math.tan(n) for n in x3]
plt.subplot(223)
plt.plot(x3,y3)
plt.suptitle('y = tan(x)')
plt.show()

# y=2^x
x4 = list(range(0,10))
y4 = [(2 ** n) for n in x4]
plt.subplot(224)
plt.plot(x4,y4)
plt.suptitle('y = 2^x')
plt.show()

plt.show()
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
# Be sure to include a legend.
```

```python
# y=√x
x = list(range(1,20))
y1 = [math.sqrt(n) for n in x]
plt.plot(x,y1,color='red', label='$sqrt$')

# y=x^3
y2 = [n ** 3 for n in x]
plt.plot(x,y2,color='blue', label='$x^3$')


# y=tan(x)
y3 = [math.tan(n) for n in x]
plt.plot(x,y3,color='orange', label='$tan(x)$')


# y=2^x
y4 = [(2 ** n) for n in x]
plt.plot(x,y4,color='black', label='$2^x$')
plt.legend()
plt.title('Visualization of all 4 functions')

plt.yscale('log')
plt.show()
```
